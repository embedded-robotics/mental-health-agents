```python
import pandas as pd
import pickle
import random

random.seed(42)
```

### Read the Combined Evaluation Responses along with Question/Answer Pairs for Validation Dataset

```python
with open('response_generation_data/all_model_responses.pkl', 'rb') as file:
    all_model_responses = pickle.load(file)

all_model_responses.head()
```

### Selecting the top 10 questions

```python
filtered_df = all_model_responses[0:10].drop(columns=['topic'])
filtered_df
```

Configuring the mapping for 10 rows

```python
original_mapping = [1,2,3,4,5]
number_of_mappings = 10
random_mappings = []

for i in range(number_of_mappings):
    random_mappings.append(random.sample(original_mapping, len(original_mapping)))

random_mappings
```

```python
mappings_archive = [[1, 5, 3, 2, 4],
                    [2, 5, 3, 1, 4],
                    [5, 4, 1, 3, 2],
                    [2, 5, 3, 1, 4],
                    [5, 4, 1, 2, 3],
                    [1, 2, 3, 4, 5],
                    [3, 2, 1, 4, 5],
                    [1, 4, 5, 2, 3],
                    [5, 3, 1, 2, 4],
                    [4, 1, 3, 2, 5]]
```

Now making a list of these randomized model indexes

```python
question_list = []
model1_ans_list = []
model2_ans_list = []
model3_ans_list = []
model4_ans_list = []
model5_ans_list = []

for index, mapping in enumerate(random_mappings):
    question_list.append(filtered_df.iloc[index, 0])
    model1_ans_list.append(filtered_df.iloc[index, mapping[0]])
    model2_ans_list.append(filtered_df.iloc[index, mapping[1]])
    model3_ans_list.append(filtered_df.iloc[index, mapping[2]])
    model4_ans_list.append(filtered_df.iloc[index, mapping[3]])
    model5_ans_list.append(filtered_df.iloc[index, mapping[4]])
    
```

```python
eval_df = pd.DataFrame({'question': question_list, 'mode1':model1_ans_list, 'model2':model2_ans_list, 'model3':model3_ans_list, 'model4':model4_ans_list, 'model5':model5_ans_list})
eval_df
```

```python
eval_df.to_csv('response_eval_data.csv', index=False)
```
